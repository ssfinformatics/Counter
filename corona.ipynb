```python
# импортируем все нужные библиотеки
import pandas as pd
import plotly.graph_objects as go
import requests
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

Загружаем данные с [сервиса](https://covid19.who.int/), который предоставляет обновляемые данные по коронавирусу 


```python
time_series = pd.read_csv('https://covid19.who.int/WHO-COVID-19-global-data.csv')
time_series['Date_reported'] = pd.to_datetime(time_series['Date_reported'])
time_series
```

```python
# сгруппируем данные по дням, суммируя случаи заражения и визуализируем при помощи plotly накопительную динамику
time_series_dates = time_series.groupby('Date_reported').sum()

fig = go.Figure()
fig.add_trace(go.Scatter(x = time_series_dates.index, y = time_series_dates['Cumulative_cases']))
fig.update_layout(title = 'Накопительная динамика случаев заражения')
fig.show()
```

```python
# кумулятивная динамика по погибшим
fig = go.Figure()
fig.add_trace(go.Scatter(x = time_series_dates.index, y = time_series_dates['Cumulative_deaths'], line_color = 'red'))
fig.update_layout(title = 'Накопительная динамика по погибшим')
```

```python
# динамика новых случаев заражения
fig = go.Figure()
fig.add_trace(go.Scatter(x = time_series_dates.index, y = time_series_dates['New_cases'], fill = 'tonexty',
                          line_color = 'gold'))
fig.update_layout(title = 'Динамика новых случаев заражения')
fig.show()
```

```python
# динамика смертей
fig = go.Figure()
fig.add_trace(go.Scatter(x = time_series_dates.index, y = time_series_dates['New_deaths'], fill = 'tonexty',
                          line_color = 'hotpink'))
fig.update_layout(title = 'Динамика количества погибших')
fig.show()
```

Рассмотрим статистику по нескольким странам


```python

# Россия
time_series_russia = time_series['Country'] == ('Russian Federation')
time_series_russia = time_series[time_series_russia]
# США 
time_series_us = time_series['Country'] == ('United States of America')
time_series_us = time_series[time_series_us]
# Бразилия
time_series_brazil = time_series['Country'] == ('Brazil')
time_series_brazil = time_series[time_series_brazil]
# Индия
time_series_india = time_series['Country'] == ('India')
time_series_india = time_series[time_series_india]
# Перу
time_series_peru = time_series['Country'] == ('Peru')
time_series_peru = time_series[time_series_peru]
```

```python
# общая динамика

fig = go.Figure()
fig.add_trace(go.Line(x = time_series_us['Date_reported'], y = time_series_us['Cumulative_cases'], name = 'USA'))
fig.add_trace(go.Line(x = time_series_brazil['Date_reported'], y = time_series_brazil['Cumulative_cases'], name = 'Brazil'))
fig.add_trace(go.Line(x = time_series_india['Date_reported'], y = time_series_india['Cumulative_cases'], name = 'India'))
fig.add_trace(go.Line(x = time_series_russia['Date_reported'], y = time_series_russia['Cumulative_cases'], name = 'Russia'))
fig.add_trace(go.Line(x = time_series_peru['Date_reported'], y = time_series_peru['Cumulative_cases'], name = 'Peru'))
fig.update_layout(title = 'Динами случаев заражения по странам')
fig.show()

# погибшие накопительно
fig = go.Figure()
fig.add_trace(go.Line(x = time_series_us['Date_reported'], y = time_series_us['Cumulative_deaths'], name = 'USA'))
fig.add_trace(go.Line(x = time_series_brazil['Date_reported'], y = time_series_brazil['Cumulative_deaths'], name = 'Brazil'))
fig.add_trace(go.Line(x = time_series_india['Date_reported'], y = time_series_india['Cumulative_deaths'], name = 'India'))
fig.add_trace(go.Line(x = time_series_russia['Date_reported'], y = time_series_russia['Cumulative_deaths'], name = 'Russia'))
fig.add_trace(go.Line(x = time_series_peru['Date_reported'], y = time_series_peru['Cumulative_deaths'], name = 'Peru'))
fig.update_layout(title = 'Динамика количества погибших')
fig.show()
```

```python
# новые заражения
fig = go.Figure()
fig.add_trace(go.Line(x = time_series_us['Date_reported'], y = time_series_us['New_cases'], name = 'USA'))
fig.add_trace(go.Line(x = time_series_brazil['Date_reported'], y = time_series_brazil['New_cases'], name = 'Brazil'))
fig.add_trace(go.Line(x = time_series_india['Date_reported'], y = time_series_india['New_cases'], name = 'India'))
fig.add_trace(go.Line(x = time_series_russia['Date_reported'], y = time_series_russia['New_cases'], name = 'Russia'))
fig.add_trace(go.Line(x = time_series_peru['Date_reported'], y = time_series_peru['New_cases'], name = 'Peru'))
fig.update_layout(title = 'Количество новых случаев заражения по дням')
fig.show()

# новые смерти
fig = go.Figure()
fig.add_trace(go.Line(x = time_series_us['Date_reported'], y = time_series_us['New_deaths'], name = 'USA'))
fig.add_trace(go.Line(x = time_series_brazil['Date_reported'], y = time_series_brazil['New_deaths'], name = 'Brazil'))
fig.add_trace(go.Line(x = time_series_india['Date_reported'], y = time_series_india['New_deaths'], name = 'India'))
fig.add_trace(go.Line(x = time_series_russia['Date_reported'], y = time_series_russia['New_deaths'], name = 'Russia'))
fig.add_trace(go.Line(x = time_series_peru['Date_reported'], y = time_series_peru['New_deaths'], name = 'Peru'))
fig.update_layout(title = 'Количество смертей по дням')
fig.show()
```

```python
# изучим данные по вакцинации
vac_data = pd.read_csv('https://covid19.who.int/who-data/vaccination-data.csv')
```

```python
# удалим пропуски
vac_data_rm = vac_data.dropna(axis=0,how="all")
vac_data_rm = vac_data.dropna(axis=0)
```

```python
full_vac = dict(
        type = 'choropleth',
        colorscale = 'Portland',
        locations = vac_data_rm['COUNTRY'],
        locationmode = "country names",
        z = vac_data_rm['PERSONS_FULLY_VACCINATED'],
        text = vac_data_rm['COUNTRY'],
        colorbar = {'title' : 'Количество вакцинированных'},
      )

layout = dict(title = 'Карт количества вакцинированных двумя дозами', 
              geo = dict(projection = {'type':'robinson'}, 
                         showlakes = True, 
                         lakecolor = 'rgb(0,191,255)'))
fig = go.Figure(data = [full_vac], 
              layout = layout)
fig.show()
```

```python
dose1_vac = dict(
        type = 'choropleth',
        colorscale = 'Portland',
        locations = vac_data_rm['COUNTRY'],
        locationmode = "country names",
        z = vac_data_rm['PERSONS_VACCINATED_1PLUS_DOSE'],
        text = vac_data_rm['COUNTRY'],
        colorbar = {'title' : 'Количество вакцинированных 1 дозой и более'},
      )

layout = dict(title = 'Карта количества вакцинированных хотя бы 1 дозой', 
              geo = dict( projection = {'type':'robinson'}, 
                         showlakes = True, 
                         lakecolor = 'rgb(0,191,255)'))
fig = go.Figure(data = [dose1_vac], 
              layout = layout)
fig.show()
```

```python
num_vac = dict(
        type = 'choropleth',
        colorscale = 'Portland',
        locations = vac_data_rm['COUNTRY'],
        locationmode = "country names",
        z = vac_data_rm['NUMBER_VACCINES_TYPES_USED'],
        text = vac_data_rm['COUNTRY'],
        colorbar = {'title' : 'Количество использованных типов вакцин'},
      )

layout = dict(title = 'Количество использованных типов вакцин по странам', 
              geo = dict( projection = {'type':'robinson'}, 
                         showlakes = True, 
                         lakecolor = 'rgb(0,191,255)'))
fig = go.Figure(data = [num_vac], 
              layout = layout)
fig.show()
```
